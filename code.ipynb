# 1. Xây dựng hàm Hash Function

## 1.1 Xây dựng class Word

```python
class Word:
    def __init__(self, en, vn):
        self.en = en
        self.vn = vn
```

## 1.2 Xây dựng class Node

```python
class Node:
    def __init__(self, en, vn):
        self.value = Word(en, vn)
        self.next = None
```

## 1.3 Xây dựng LinkedList

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def print(self):
        if self.head is not None:
            tmp = self.head
            while (tmp.next is not None):
                print(tmp.value.en,':',tmp.value.vn)
                tmp = tmp.next
            print(tmp.value.en,':', tmp.value.vn)

    def add_last(self, value):  #('two ', ' hai') => two -> hai, 2
        newnode = Node(value[0], value[1])
        if self.head is None:   
            self.head = newnode # đối tượng của word
            return True
        else:
            last = self.head
            if newnode.value.en == last.value.en:
                return False
            while (last.next):
                # check trùng
                last = last.next
                if newnode.value.en == last.value.en:
                    return False        
            last.next = newnode # đối tượng của word
            return True
                
    def remove(self, value): # ok
        # Store head node
        temp = self.head
 
        # If head node itself holds the key to be deleted
        if (temp is not None):
            if (temp.value.en.strip() == value):
                self.head = temp.next
                temp = None
                return True
        # Search for the key to be deleted, keep track of the previous node as we need to change 'prev.next'
        while(temp is not None):
            if temp.value.en.strip() == value:
                break
            prev = temp
            temp = temp.next
 
        # if key was not present in linked list
        if(temp == None):
            return False
        # Unlink the node from linked list
        prev.next = temp.next
        temp = None
        return True
        
    def find(self, value):
        if self.head is not None:
            index = 0
            node = self.head
            while node is not None:
                if node.value.en.strip() == value:
                    return index, node.value.vn
                node = node.next
                index += 1
        return None
        
    def get_size(self):
        count = 0
        if self.head is not None:
            node = self.head
            while node is not None:
                node = node.next
                count += 1
        return count
```

## 1.4 Xây dựng HashTable

```python
class HashTable:
    def __init__(self, M):
        self.hashTable = [LinkedList() for _ in range(M)]
        self.M = M

    def h(self, x):
        # ord(): char -> ASCII 
        # chr(): ASCII  -> char
        key_ascii = sum([ord(x[i])*(i+1) for i in range(len(x))])

        return key_ascii % self.M

    def insert(self, x):
        en, vn = (x[0].lower().strip(), x[1].lower().strip())
        index = self.h(en)
        if self.hashTable[index].add_last(tuple([en, vn])):
            return True
        else:
            return False

    def delete(self, x): 
        index = self.h(x.lower().strip())
        if self.hashTable[index].remove(x):
            return True
        else:
            return False

    def get(self, x): 
        index = self.h(x)
        index_vn = self.hashTable[index].find(x)
        if index_vn is not None:
            return (index, index_vn[0])
        else:
            return None

    def means(self, x): 
        index = self.h(x)
        index_vn = self.hashTable[index].find(x)
        if index_vn is not None:
            return index_vn[1]
        else:
            return None
    
    def showHashTable(self):
        for i in range(self.M):
            if self.hashTable[i].head is not None:
                print('======Index '+str(i) + ' hashtable======')
                self.hashTable[i].print()

    def hash_size(self):
        count = 0
        for i in range(self.M):
            count += self.hashTable[i].get_size()
        return count

    def get_longest(self):
        index = 0
        max_length = 0
        for i in range(self.M):
            if max_length < self.hashTable[i].get_size():
                max_length = self.hashTable[i].get_size()
                index = i
        return index

    def size_longest(self):
        return self.hashTable[self.get_longest()].get_size()

    def show_longest(self):
        self.hashTable[self.get_longest()].print()
```

# 2. Khởi tạo từ điển Hash

```python
# create hash function
M = 199
hash = HashTable(M)
```

# 3. Thêm từ mới

```python
# load dữ liệu
with open('data.txt','r',encoding='utf-8') as f:
    data = f.read().split('\n')
data[:5]
```

```python
# insert values
for en_vn in data:
    key_val = en_vn.split(' | ')  # ['map ', ' bản đồ']
    hash.insert(key_val)
hash.showHashTable()
```

```python
# từ này có rồi nên sẽ ra False
print(hash.insert(['system', 'hệ thống']))
```

```python
# 2 từ này mới nên ra True
print(hash.insert(['hello', 'xin chào']))
print(hash.insert(['cry', 'khóc']))
hash.showHashTable()
```
